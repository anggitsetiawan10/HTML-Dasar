```python
!pip install scikit-learn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
# Asumsikan 'dataset' adalah DataFrame pandas Anda

# Ekstrak fitur dan variabel target
x = dataset.iloc[:, :-1]  # Pertahankan sebagai DataFrame untuk OneHotEncoder
y = dataset.iloc[:, -1].values

# Bagi data menjadi set pelatihan dan pengujian
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Buat objek OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False untuk array numpy

# Sesuaikan encoder pada data pelatihan dan transformasikan data pelatihan dan pengujian
x_train_encoded = encoder.fit_transform(x_train[['Outlook', 'Temperature','Humidity','Windy']]) # Asumsikan 'Outlook' adalah kolom kategorikal Anda dengan 'overcast', 'sunny'
x_test_encoded = encoder.transform(x_test[['Outlook', 'Temperature','Humidity','Windy']])

# Buat objek StandardScaler
sc = StandardScaler()

# Sesuaikan scaler pada data pelatihan yang dikodekan dan transformasikan data pelatihan dan pengujian
x_train = sc.fit_transform(x_train_encoded)
x_test = sc.transform(x_test_encoded)
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
classifier.predict_proba(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Impor fungsi yang diperlukan
from sklearn.metrics import classification_report

# Sekarang Anda dapat menggunakan fungsi classification_report
akurasi = classification_report(y_test, y_pred)
print(akurasi)
```

```python
akurasi = accuracy_score(y_test, y_pred)
print("Tingkat akurasi : %d persen"%(akurasi*100))
```
